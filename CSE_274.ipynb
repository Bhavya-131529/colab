```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train.shape, X_test.shape
```

.fit() ---->calculates mean and std from training only

transform() ---->applies learned parameters

LOW VARIANCE THRESHOLD

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selector = VarianceThreshold(0.01)
X_train_var = selector.fit_transform(X_train)
X_test_var = selector.transform(X_test)
X_train_var.shape, X_test_var.shape
```

```python
pd.DataFrame(X_train, columns=X.columns).corr().abs()
```

```python
corr_matrix=pd.DataFrame(X_train,columns=X.columns).corr().abs()
upper=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
to_drop=[column for column in upper.columns if any(upper[column]>0.9)]
to_drop
```

```python
X_train_corr=pd.DataFrame(X_train,columns=X.columns).drop(columns=to_drop)
X_test_corr=pd.DataFrame(X_test,columns=X.columns).drop(columns=to_drop)
X_train_corr.shape, X_test_corr.shape
```

Tree based feature importance

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train_corr,y_train)
```

```python
importances=pd.Series(rf.feature_importances_,index=X_train_corr.columns)
importances.sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,6))
importances.plot(kind="barh")
plt.gca().invert_yaxis()
plt.title("Feature Importance (All 20 features)")
plt.xlabel("Importance")
plt.show()
```

#PCA (Principal Component Analysis)
> Onlt after correlation and is used for doing the dimensionality reduction.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(X_train_corr)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
pca =PCA(n_components=0.95)
X_train_pca=pca.fit_transform(X_train_corr)
X_test_pca=pca.transform(X_test_corr)
X_train_pca.shape, X_test_pca.shape
```
